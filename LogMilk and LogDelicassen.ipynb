```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import HoverTool
import math
output_notebook()
```

```python
import xlrd
file_location = "C:/Users/Sandeep/Desktop/CollegeCourses/GRAD_SCHOOL/CS_690V_Visual_Analytics/HW/HW2/Wholesale customers data.xlsx"
with6ValuesFile = xlrd.open_workbook(file_location)
with6Vals = with6ValuesFile.sheet_by_index(0)

file_location = "C:/Users/Sandeep/Desktop/CollegeCourses/GRAD_SCHOOL/CS_690V_Visual_Analytics/HW/HW2/Wholesale customers data-missing.xlsx"
without6ValuesFile = xlrd.open_workbook(file_location)
without6Vals = without6ValuesFile.sheet_by_index(0)
```

```python
wholeDataMilk = []

for r in range(1, with6Vals.nrows):
    wholeDataMilk.append(with6Vals.cell_value(r,3))

#print(wholeDataMilk)
#print(len(wholeDataMilk))
```

```python
wholeDataDelicassen = []

for r in range(1, with6Vals.nrows):
    wholeDataDelicassen.append(with6Vals.cell_value(r,7))
        
#print(wholeDataDelicassen)
#print(len(wholeDataDelicassen))
```

```python
missingDataDelicassen = []

for r in range(1, without6Vals.nrows):
    
    if without6Vals.cell_value(r,7) == '':
        missingDataDelicassen.append("EMPTY SLOT")
    else:
        missingDataDelicassen.append(without6Vals.cell_value(r,7))

print(missingDataDelicassen)
print(len(missingDataDelicassen))
```

```python
avgDelicassen = 0

for i in missingDataDelicassen:
    if i != "EMPTY SLOT":
        avgDelicassen+=i

avgDelicassen = avgDelicassen / (len(missingDataDelicassen)-2)
print(avgDelicassen)
```

```python
missingIndices = []

for i in range(0, 440):
    if missingDataDelicassen[i] == "EMPTY SLOT":
        missingDataDelicassen[i] = avgDelicassen
        missingIndices.append(i)
        
print(missingIndices)
```

```python
log10Milk = []

for i in wholeDataMilk:
    log10Milk.append(math.log10(i))
```

```python
log10Delicassen = []

for i in missingDataDelicassen:
    log10Delicassen.append(math.log10(i))
```

```python
p = figure(plot_width=400, plot_height=400, title="US Employment Rates after 1991")

p.circle(log10Milk, log10Delicassen, line_color="navy", fill_color="orange", fill_alpha=0.5)

p.xaxis.axis_label = "Log10(Milk)"
p.yaxis.axis_label = "Log10(Delicassen)"

show(p)
```

```python
avgLog10Milk = 0

for i in log10Milk:
    avgLog10Milk+=i
    
avgLog10Milk = avgLog10Milk / len(log10Milk)

print(avgLog10Milk)
```

```python
avgLog10Delicassen = 0

for i in log10Delicassen:
    avgLog10Delicassen+=i
    
avgLog10Delicassen = avgLog10Delicassen / len(log10Delicassen)

print(avgLog10Delicassen)
```

```python
log10MilkVSlog10DelicassenSlopeNumerator = 0
log10MilkVSlog10DelicassenSlopeDenominator = 0

for i in range(0, 439):
    log10MilkVSlog10DelicassenSlopeNumerator += (log10Milk[i] - avgLog10Milk)*(log10Delicassen[i] - avgLog10Delicassen)
    log10MilkVSlog10DelicassenSlopeDenominator += (log10Milk[i] - avgLog10Milk)*(log10Milk[i] - avgLog10Milk)

log10MilkVSlog10DelicassenSlope = log10MilkVSlog10DelicassenSlopeNumerator / log10MilkVSlog10DelicassenSlopeDenominator
print("Slope for Log10(Milk) vs Log10Delicassen is = ",log10MilkVSlog10DelicassenSlope)
```

```python
log10MilkVSlog10DelicassenIntercept = log10Delicassen[220] - (log10MilkVSlog10DelicassenSlope*log10Milk[220])
print("Intercept for Log10(Milk) vs Log10Delicassen is = ", log10MilkVSlog10DelicassenIntercept)
```

```python
def findLog10Delicassen(log10MilkNum):
    log10DelicassenEstimate = log10MilkVSlog10DelicassenIntercept + (log10MilkVSlog10DelicassenSlope*log10MilkNum)
    return log10DelicassenEstimate
```

```python
log10DelicassenPredictions = []

for i in log10Milk:
    log10DelicassenPredictions.append(findLog10Delicassen(i))

delicassenPredictions = []
missingValuePredictions = []
correctValues= []

for i in range(0, 440):#log10DelicassenPredictions:
    if i in missingIndices:
        missingValuePredictions.append(math.pow(log10DelicassenPredictions[i], 10))
        correctValues.append(wholeDataDelicassen[i])
        
    delicassenPredictions.append(math.pow(log10DelicassenPredictions[i], 10))
    
print(missingValuePredictions)
print(correctValues)
```

```python
MilkVSDelicassenVSDelicassenPredictionsMSE = 0

for i in range(0, 439):
    diffSquared = math.pow(delicassenPredictions[i] - wholeDataDelicassen[i], 2)
    MilkVSDelicassenVSDelicassenPredictionsMSE += diffSquared

MilkVSDelicassenVSDelicassenPredictionsMSE = MilkVSDelicassenVSDelicassenPredictionsMSE / 440
print("MSE for Log10(Milk) vs Log10(Delicassen) is = ", MilkVSDelicassenVSDelicassenPredictionsMSE)
```

```python
p = figure(plot_width=400, plot_height=400, title="US Employment Rates after 1991")

p.circle(wholeDataMilk, delicassenPredictions, line_color="navy", fill_color="orange", fill_alpha=0.5)

p.xaxis.axis_label = "Milk"
p.yaxis.axis_label = "Delicassen Prediction"

show(p)
```
